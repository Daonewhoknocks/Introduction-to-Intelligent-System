```python
dataset = pd.read_csv("Lesson_4_student_grades_dataset.csv")
dataset.head()
print(dataset.isnull().sum())
```

3. Split Text and Training

4. Linear Regression model

```python
x = dataset.drop('Grades', axis = 1)
y = dataset['Grades']
```

5. Model Training

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train,y_train)
```

6. Prediction and Evaluiation

```python
new_data = np.array([[5, 90, 80, 1, 30]])
prediction = model.predict(new_data)

print("Predicted Grade:", prediction[0])
```

7. Features importance analysis

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
x = 'Previous_Exam_Scores'
y = 'Grades'

print(data.head())
```

```python
plt.scatter(data[x],data[y])
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Varialble')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
x = data[[x]]
y = data[[y]]
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope", slope)
print("intercept", intercept)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), color='red')
plt.xlabel('Independent Variable')
plt.ylabel('dependent Variable')
plt.title('Linear Regression')
plt.show()
```

```python
new_x = [[np.random.randint(100,200)]]
predict_y = model.predict(new_x)
print("New X:", new_x)
print("Predict Y from new X:", predict_y)
```
