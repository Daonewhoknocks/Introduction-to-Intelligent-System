```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
warnings.filterwarnings('ignore')
```

1. step 1

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
x = 'Hours_of_Study'
y = 'Grades'

print(data.head())
```

2. step2

```python
plt.scatter(data[x],data[y])
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Varialble')
plt.title('Scatter Plot of Data')
plt.grid()
plt.show()
```

step3

```python
x = data[[x]]
y = data[[y]]
```

step4

```python
model = LinearRegression()
model.fit(x,y)
```

step5

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope", slope)
print("intercept", intercept)
```

step6

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), color='red')
plt.xlabel('Independent Variable')
plt.ylabel('dependent Variable')
plt.title('Linear Regression')
plt.grid()
plt.show()
```

step7

```python
new_x = [[np.random.randint(100,200)]]
predict_y = model.predict(new_x)
print("new x:", new_x)
print("Predict Y from new X:", predict_y)
```
