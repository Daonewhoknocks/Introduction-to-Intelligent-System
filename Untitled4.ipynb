SIMPLE LINEAR REGRESSION

1. IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np #for numerical
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

2. create date frame

```python
num_point=50
x = np.random.randint(0,100,num_point)
y = 2 * x + np.random.normal(0,10,num_point)

data = pd.DataFrame({'X': x, 'Y': y})
data.to_csv('data.csv', index=False)
print(data.head())
```

3. Visualize the data

```python
plt.scatter(data['X'],data['Y'])
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Varialble')
plt.title('Scatter Plot of Data')
plt.show()
```

4. Prepatation for modeling

```python
x = data[['X']]
y = data[['Y']]
```

5. fitting the linear

```python
model = LinearRegression()
model.fit(x,y)
```

6. coefficient

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope", slope)
print("intercept", intercept)
```

7. visualizing the linear

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), color='red')
plt.xlabel('Independent Variable')
plt.ylabel('dependent Variable')
plt.title('Linear Regression')
plt.show()
```

8. making prediction

```python
new_x = [[np.random.randint(100,200)]]
predict_y = model.predict(new_x)
print("Predict Y from new X:", predict_y)
```
